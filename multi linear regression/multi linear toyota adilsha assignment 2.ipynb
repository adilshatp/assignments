```python
methods=(('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight'),
         ('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax'),
         ('Price~Age_08_04+KM+HP+cc+Doors+Gears+Weight'),
         ('Price~Age_08_04+KM+HP+cc+Doors+Quarterly_Tax+Weight'),
         ('Price~Age_08_04+KM+HP+cc+Gears+Quarterly_Tax+Weight'),
         ('Price~Age_08_04+KM+HP+Doors+Gears+Quarterly_Tax+Weight'),
         ('Price~Age_08_04+KM+cc+Doors+Gears+Quarterly_Tax+Weight'),
         ('Price~Age_08_04+HP+cc+Doors+Gears+Quarterly_Tax+Weight'),
         ('Price~KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight'),('Price ~ Age_08_04'),
         ('Price ~ KM'),('Price ~ HP'),('Price ~ cc'),
         ('Price ~ Gears'),('Price ~ Quarterly_Tax'),('Price ~ Weight'))
```

```python
import statsmodels.formula.api as smf
```

```python
for model in methods:
    models2=smf.ols(model,data=scaled_dfnew)
    models1=models2.fit()
    params=models1.params
    print(params)
    pvalue=models1.pvalues
    pvalue_data=pvalue.to_frame()
    pvalue_data.reset_index(inplace=True)
    pvalue_data.rename(columns={'index':'features'},inplace=True)
    print('')
    print(pvalue_data)
    for i in range(len(pvalue_data)):
        if i!=0:
            if pvalue_data.iloc[i,1]>0.05:
                print('')
                print('{} is a insignificant feature , since pvalue {} is greater than 0.5'.format(pvalue_data.iloc[i:0],pvalue_data.iloc[i:1]))
    print('')
    print('aic value:',models1.aic)
    print('rsquare:',models1.rsquared) 
    print('*___________________________________________________________*')
```

In most of the cases, insignificant/ non-important feature is Door, there contribution in model buliding is comparitively very less than other feature fallowed by cc
The model with comparitively less aic value and more r2 value is considered as a best model than other model

The model built by taking all the feature and the one built by dropping Door performes better than other model

#  vif

```python
req_age=smf.ols('Age_08_04~KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=scaled_dfnew).fit().rsquared
vif_age=1/(1-req_age)

req_km=smf.ols('KM~Age_08_04+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=scaled_dfnew).fit().rsquared
vif_km=1/(1-req_km)

req_hp=smf.ols('HP~Age_08_04+KM+cc+Doors+Gears+Quarterly_Tax+Weight',data=scaled_dfnew).fit().rsquared
vif_hp=1/(1-req_hp)

req_cc=smf.ols('cc~Age_08_04+KM+HP+Doors+Gears+Quarterly_Tax+Weight',data=scaled_dfnew).fit().rsquared
vif_cc=1/(1-req_cc)

req_doors=smf.ols('Doors~Age_08_04+KM+HP+cc+Gears+Quarterly_Tax+Weight',data=scaled_dfnew).fit().rsquared
vif_door=1/(1-req_doors)

req_gears=smf.ols('Gears~Age_08_04+KM+HP+cc+Doors+Quarterly_Tax+Weight',data=scaled_dfnew).fit().rsquared
vif_gear=1/(1-req_gears)

req_qt=smf.ols('Quarterly_Tax~Age_08_04+KM+HP+cc+Doors+Gears+Weight',data=scaled_dfnew).fit().rsquared
vif_qt=1/(1-req_qt)

req_weight=smf.ols('Weight~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax',data=scaled_dfnew).fit().rsquared
vif_wt=1/(1-req_weight)
```

```python
vif_data=pd.DataFrame({'variable':['age','km','hp','cc','door','gear','quarterly','wt'],
                      'VIF':[vif_age,vif_km,vif_hp,vif_cc,vif_door,vif_gear,vif_qt,vif_wt]})
```

## residual plot for homscedasticity

```python
scalar=StandardScaler()
```

```python
residqq=pd.DataFrame(modelqq.resid)
fittedqq=pd.DataFrame(modelqq.fittedvalues)
```

```python
residqq1=scalar.fit_transform(residqq)
fittedqq1=scalar.fit_transform(fittedqq)
```

```python
plt.scatter(fittedqq1,residqq1)
plt.title('residual plot for homscedasticity')
plt.xlabel('fitted value')
plt.ylabel('fitted residual')
plt.show()
```
