```python
movies.drop('V2',axis=1,inplace=True)
movies.drop('V3',axis=1,inplace=True)
movies.drop('V5',axis=1,inplace=True)
movies.drop('V4',axis=1,inplace=True)
```

```python
movies
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
f_items1=apriori(df=movies,min_support=0.2,use_colnames=True)
f_items1
```

```python
a_rules=association_rules(df=f_items1,metric='support',min_threshold=0.05)
a_rules
```

```python
supports=[0.1,0.3,0.5,0.7]
thresholds=[0.2,0.5,0.7]

for i in supports:
    f_items1=apriori(movies,min_support=i,use_colnames=True)
    print("support ia :",i)
    print('f_items:',f_items1)
    print('----------------------')
    for t in thresholds:
        a_rules=association_rules(f_items1,min_threshold=t)
        print('threshold is:',t)
        print(a_rules)
```

```python
f_items2=apriori(df=movies,min_support=0.05,use_colnames=True)
f_items2
```

```python
a_rules_n=association_rules(df=f_items2,metric='support',min_threshold=0.05)
a_rules_n
```

```python
plt.scatter(x=a_rules_n['support'],y=a_rules_n['confidence'])
```

```python
a_rules_n=association_rules(df=f_items2,metric='lift',min_threshold=0.01)
a_rules_n
```

```python
plt.scatter(x=a_rules_n['lift'],y=a_rules_n['confidence'])
```
