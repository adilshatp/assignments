```python
from sklearn.preprocessing import StandardScaler
crime_std=StandardScaler().fit_transform(crime1)
```

```python
dendogram=sch.dendrogram(sch.linkage(crime_scaled,method='single',metric='euclidean'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
h_c=AgglomerativeClustering(n_clusters=2,linkage='single')
```

```python
y_h_c=h_c.fit_predict(crime_scaled)
y_h_c
```

```python
cluster_age=pd.DataFrame({'cluster':y_h_c})
cluster_age.head()
```

```python
crime_scaled['cluster']=cluster_age
```

```python
# k mean
```

```python
from sklearn.cluster import KMeans
```

```python
x=[]
for i in range (1,11):
    cluster_kmean=KMeans(n_clusters=i)
    cluster_kmean.fit(crime_std)
    x.append(cluster_kmean.inertia_)
    
plt.plot(range(1,11),x)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

There is a deap dip in the elbow curve which gives the optimum number of K-values

When the value of K is 2, the dip is more than others

Hence, optimum number of K value is 2

```python
cluster_kmean=KMeans(n_clusters=2)
cluster_kmean.fit(crime_std)
```

```python
cc=cluster_kmean.labels_
```

```python
crime2=crime
```

```python
crime2['cluster']=cc
```
