```python
from sklearn.preprocessing import StandardScaler
```

```python
std=StandardScaler()
```

```python
airline_std=std.fit_transform(airline)
```

```python
airline_std
```

```python
airline_scaled=pd.DataFrame(airline_std,columns=['Balance','Qual_miles','cc1_miles','cc2_miles','cc3_miles','Bonus_miles','Bonus_trans','Flight_miles_12mo','Flight_trans_12','Days_since_enroll','Award?'])
```

```python
airline_scaled
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendogram=sch.dendrogram(sch.linkage(airline_scaled,method='single',metric='euclidean'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
h_c=AgglomerativeClustering(n_clusters=2,linkage='single')
```

```python
y_h_c=h_c.fit_predict(airline_scaled)
y_h_c
```

```python
airline1=airline
```

```python
airline1['cluster']=pd.DataFrame(y_h_c)
```

```python
airline1
```

## kmean

```python
from sklearn.cluster import KMeans
```

```python
x=[]
for i in range(1,11):
    k_airline=KMeans(n_clusters=i,random_state=12)
    k_airline.fit(airline_scaled)
    x.append(k_airline.inertia_)
```

```python
plt.plot(range(1,11),x)
plt.title('elbowchart')
plt.xlabel('squarerd error')
plt.ylabel('no of clusters')
```

```python
k_airlines=KMeans(n_clusters=4,random_state=12)
k_airlines.fit(airline_scaled)
```

```python
kk=k_airlines.labels_
```

```python
airline2=airline
```

```python
airline2['cluster']=kk
```

```python
airline2
```

```python
airline2['cluster'].nunique()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
cluster=d_airline.labels_
cluster
```
