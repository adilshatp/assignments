```python
lab=lab.drop(['age','educationno','capitalgain','capitalloss','hoursperweek'])
```

```python
lab
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in lab:
    salary_train[i]=le.fit_transform(salary_train[i])
```

```python
salary_train
```

```python
for i in lab:
    salary_test[i]=le.fit_transform(salary_test[i])
```

```python
salary_test
```

```python
x_train=salary_train.drop('Salary',axis=1)
```

```python
y_train=salary_train['Salary']
```

```python
x_train.shape,y_train.shape
```

```python
x_test=salary_test.drop('Salary',axis=1)
```

```python
y_test=salary_test['Salary']
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled=StandardScaler()
```

```python
scaled_df=scaled.fit_transform(x_train)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nn=GaussianNB()
```

```python
nn.fit(scaled_df,y_train)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
yp=nn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
accuracy_score(y_test,yp)
```

```python
print(classification_report(y_test,yp))
```

```python
cm=confusion_matrix(y_test,yp)
cm
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nn2=MultinomialNB()
```

```python
nn2.fit(x_train,y_train)
```

```python
yp2=nn2.predict(x_test)
```

```python
accuracy_score(y_test,yp2)
```
