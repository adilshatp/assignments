### using cosine system

```python
books_cosine=pd.pivot_table(data=books,values='Book.Rating',index='User.ID',columns='Book.Title').fillna(0)
books_cosine.index=books["User.ID"].unique()
books_cosine
```

```python
cosine_df=1 - pairwise_distances(X=books_cosine,metric='cosine')
cosine_data=pd.DataFrame(data=cosine_df)
cosine_data.index=books["User.ID"].unique()
cosine_data.columns=books["User.ID"].unique()
cosine_data
```

```python
import numpy as np
np.fill_diagonal(a=cosine_data.to_numpy(),val= 0)
```
