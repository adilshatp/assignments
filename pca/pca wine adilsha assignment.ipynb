# model building

```python
values=df1.values
values
```

```python
type(values)
```

```python
from sklearn.preprocessing import scale
```

```python
std_values=scale(values)
```

```python
std_values
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)
```

```python
pca_scaled=pca.fit_transform(std_values)
pca_scaled
```

```python
varience=pca.explained_variance_ratio_
varience
```

```python
varience2=np.cumsum(np.round(varience,decimals=3)*100)
varience2
```

```python
pca_scaled.shape
```

```python
pca.components_
```

```python
plt.plot(varience2,color='red')
```

```python
final = pd.concat([df["Type"],pd.DataFrame(pca_scaled,columns=["pc1","pc2","pc3"])],axis=1)
```

```python
final.head()
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.scatterplot("pc1","pc2",data=final,hue="Type")
```

# scree plot

```python
pc_values=np.arange(pca.n_components_) + 1
plt.plot(pc_values,pca.explained_variance_ratio_,'o-',linewidth=2,color='blue')
plt.title('scree plot')
plt.xlabel('principal component')
plt.ylabel('varience explained')
plt.show()
```
