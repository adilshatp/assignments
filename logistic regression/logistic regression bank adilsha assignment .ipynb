# data preparation

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
bank[['job','marital','education','default','balance','housing','loan','contact','day','month','duration','campaign','pdays','previous','poutcome']].nunique()
```

```python
bank['job']=lb.fit_transform(bank[['job']])
bank['marital']=lb.fit_transform(bank[['marital']])
bank['education']=lb.fit_transform(bank[['education']])
bank['default']=lb.fit_transform(bank[['default']])
bank['balance']=lb.fit_transform(bank[['balance']])
bank['housing']=lb.fit_transform(bank[['housing']])
bank['loan']=lb.fit_transform(bank[['loan']])
bank['contact']=lb.fit_transform(bank[['contact']])
bank['day']=lb.fit_transform(bank[['day']])
bank['month']=lb.fit_transform(bank[['month']])
bank['duration']=lb.fit_transform(bank[['duration']])
bank['campaign']=lb.fit_transform(bank[['campaign']])
bank['pdays']=lb.fit_transform(bank[['pdays']])
bank['previous']=lb.fit_transform(bank[['previous']])
bank['poutcome']=lb.fit_transform(bank[['poutcome']])
bank['y']=lb.fit_transform(bank[['y']])
```

```python
bank.head()
```

```python
bank.dtypes
```

# model building

```python
x=bank.iloc[:,:16]
y=bank['y']
```

```python
x.head()
```

```python
y.head()
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=12)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

# model training

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

# model testing

```python
yp_train=lr.predict(x_train)
yp_test=lr.predict(x_test)
```

# model evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score,auc
```

```python
print(classification_report(yp_train,y_train))
```

```python
print(classification_report(yp_test,y_test))
```

```python
print(confusion_matrix(yp_train,y_train))
```

```python
print(confusion_matrix(yp_test,y_test))
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
fpr,tpr,threshold=roc_curve(y,lr.predict_proba (x)[:,1])
auc=roc_auc_score(y_test,yp_test)
print(auc)
```

```python
import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logic model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```
